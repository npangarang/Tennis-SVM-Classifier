```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
pd.set_option('display.max_columns', 30)
```

```python
data = pd.read_csv('Final_data.csv').drop(['Unnamed: 0'], axis = 1)
roundReference = pd.read_csv('Round_reference.csv').set_index('round')
surfaceReference = pd.read_csv('Surface_reference.csv').set_index('surface')
tourneyReference = pd.read_csv('Tourney_reference.csv').set_index('tourney_name')
countryReference = pd.read_csv('Country_reference.csv').set_index('winner_ioc')

display(data.head())
display(len(data))
display(roundReference.head())
display(surfaceReference.head())
display(tourneyReference.head())
display(countryReference.head())
```

```python
#Function that returns a dataframe of all the previous matches of two given players
def getRecord(p1, p2):
    history = data[:0]
    for x in range(len(data)):
        row = data.iloc[x]
        if p1 == row['p1'] and p2 == row['p2']:
            history = history.append(row)
        elif p2 == row['p1'] and p1 == row['p2']: 
            history = history.append(row)
    history = history.reset_index().drop(['index'], axis = 1)
    return history
```

```python
fedal = getRecord('Roger Federer', 'Rafael Nadal')
display(fedal.head())
display(len(fedal))
```

```python
#This function will assign p1 to 1, p2 to 0, and winner to 1 and standardize the data
#This function will also get rid of p1 and p2 so that the p1 isn't the obvious winner
def normalize(data, p1, p2):
    
    #data.loc[data.p1 == p1, 'p1'] = 1 #p1
    #data.loc[data.p1 == p2, 'p1'] = 0
    #data.loc[data.p2 == p1, 'p2'] = 1 #p2
    #data.loc[data.p2 == p2, 'p2'] = 0
    data.loc[data.winner == p1, 'winner'] = 1 #p1
    data.loc[data.winner == p2, 'winner'] = 0
    data = data.drop(['p1'], axis = 1)
    data = data.drop(['p2'], axis = 1)
    
    columns = data.columns.tolist()
    columns.remove('winner')
    scaler = MinMaxScaler(feature_range=(0,1))
    for col in columns:
        data[col] = scaler.fit_transform(data[col].values.reshape(-1, 1))
    return data
```

```python
#Normalized Data
fedalNorm = normalize(fedal, 'Roger Federer', 'Rafael Nadal')
display(fedalNorm.head(10))
display(len(fedalNorm))
```

The previous dataframe represents a standardized dataframe of the matches played between Federer and Nadal. The problem with this is that my classifier would struggle to analyze a matchup in which both players have played very few matches against each other, or if they haven't played, as there would be very little to no data to utilize. In this dataframe of Federer and Nadal's matches alone, we only see 23 matches which is still a minimal amount of data to run algorithms on.

Because of this, I tried training the full 27,000-row-long original data frame after label encoding each individual player only to realize that I have ended up with too many possible winners to predict based on the data. 

```python
#Back to data preprocessing
from sklearn.preprocessing import LabelEncoder

p = data['p1'].drop_duplicates()
categorical_cols = ['p1']
le = LabelEncoder()
data[categorical_cols] = data[categorical_cols].apply(lambda col: le.fit_transform(col))
pp = data['p1'].drop_duplicates().rename('value')
playerReference = pd.concat([p,pp], axis = 1).set_index('p1').reindex()
display(playerReference.head())
display(data.head())
#Normalizing original dataframe
#columns = data.columns.tolist()
#columns.remove('winner')
#scaler = MinMaxScaler(feature_range=(0,1))
#for col in columns:
#    data[col] = scaler.fit_transform(data[col].values.reshape(-1, 1))
```

```python
#Support Vector Machine
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(fedalNorm, np.ravel(fedalNorm.winner), test_size=0.30, random_state=101)

from sklearn.svm import SVC
model = SVC()
model.fit(X_train,y_train)

predictions = model.predict(X_test)
```

Predictions

```python
predictions = model.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001]} 
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=2)
grid.fit(X_train,y_train)
grid_predictions = grid.predict(X_test)
```

```python
print(confusion_matrix(y_test,grid_predictions))
```

```python
print(classification_report(y_test,grid_predictions))
```
