# Filtering Data

```python
#Uneeccesary columns 
df = df.drop(['tourney_id', 'draw_size', 'tourney_date', 'match_num', 'winner_id', 'winner_seed', 'winner_entry', 'winner_age', 'loser_id', 'loser_seed', 'loser_entry', 'loser_age', 'winner_rank_points', 'loser_rank_points', 'winner_ht', 'loser_ht', 'score', 'minutes'], axis = 1)
#Filter tournament levels to only A/M/G level tournaments
df = df[(df['tourney_level'] == 'A') | (df['tourney_level'] == 'M') | (df['tourney_level'] == 'G')]
df.index = np.arange(0,len(df))
df.head()
```

# Encoding data/Feature Engineering

```python
#Label encoding categorical data
dfOld = df.copy() #copy of data before pre-encoding
from sklearn.preprocessing import LabelEncoder

categorical_feature_mask = df.dtypes==object
categorical_cols = df.columns[categorical_feature_mask].tolist()

le = LabelEncoder()
df[categorical_cols] = df[categorical_cols].apply(lambda col: le.fit_transform(col))
```

```python
#dictionary to refer categorical columns later
ref = {'tourney_name': {}, 'surface': {}, 'tourney_level' : {}, 'winner_name' : {}, 'winner_hand' 
       : {}, 'winner_ioc' : {}, 'loser_name' : {}, 'loser_hand' : {}, 'loser_ioc' : {}, 'round' : {}}
for (cat1, cat2) in zip(df, dfOld):
    for i in range(len(df)):
        ref['tourney_name'][df['tourney_name'][i]] = dfOld['tourney_name'][i]
        ref['surface'][df['surface'][i]] = dfOld['surface'][i]
        ref['tourney_level'][df['tourney_level'][i]] = dfOld['tourney_level'][i]
        ref['winner_name'][df['winner_name'][i]] = dfOld['winner_name'][i]
        ref['winner_hand'][df['winner_hand'][i]] = dfOld['winner_hand'][i]
        ref['winner_ioc'][df['winner_ioc'][i]] = dfOld['winner_ioc'][i]
        ref['loser_name'][df['loser_name'][i]] = dfOld['loser_name'][i]
        ref['loser_hand'][df['loser_hand'][i]] = dfOld['loser_hand'][i]
        ref['loser_ioc'][df['loser_ioc'][i]] = dfOld['loser_ioc'][i]
        ref['round'][df['round'][i]] = dfOld['round'][i]
```

```python
#Engineering player 1 / player 2 / winner feature columns
df = df.rename(columns={'winner_name' : 'p1', 'loser_name' : 'p2'})
df.head()
```

```python
#reading reference dictionary and dataframe to csv files
import csv
with open('reference.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in ref.items():
       writer.writerow([key, value])
df.to_csv()
```
