```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
pd.set_option('display.max_columns', 30)
```

```python
data = pd.read_csv('Final_data.csv').drop(['Unnamed: 0'], axis = 1)
roundReference = pd.read_csv('Round_reference.csv').set_index('round')
surfaceReference = pd.read_csv('Surface_reference.csv').set_index('surface')
tourneyReference = pd.read_csv('Tourney_reference.csv').set_index('tourney_name')
countryReference = pd.read_csv('Country_reference.csv').set_index('winner_ioc')
playerReference = pd.read_csv('Player_reference.csv').set_index('winner_name')

display(data.head())
display(len(data))
display(roundReference.head())
display(surfaceReference.head())
display(tourneyReference.head())
display(countryReference.head())
display(playerReference.head())
```

```python
#Now, let's normalize the data so we can feed it into our machine learning algos
columns = data.columns.tolist()
scaler = MinMaxScaler(feature_range=(0,1))
for col in columns:
    data[col] = scaler.fit_transform(data[col].values.reshape(-1, 1))
#We should also drop the p1 p2 columnss
data.head()
```

```python
woh = data.head(50)
len(woh)
```

```python
#Support Vector Machine
from sklearn.model_selection import train_test_split
X = woh.drop('winner',axis=1)
y = woh['winner']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)

from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train,y_train)
```

```python
predictions = svc_model.predict(X_test)
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
