# Filtering Data

```python
df = df.drop(['tourney_id', 'draw_size', 'tourney_date', 'match_num', 'winner_id', 'winner_entry', 'loser_id', 'loser_entry', 'winner_rank_points', 'loser_rank_points', 'winner_seed', 'loser_seed', 'best_of'], axis = 1)
#I'm going to drop all the rows with NaN values.
#This is probably not the best way to handle NaN values, but oh well ¯\_(ツ)_/¯
df = df.dropna()
#Also dropping Davis Cup Matches
df = df[df.tourney_level != 'D']
```

```python
display(df.head())
display(len(df))
```

Now let's find quantify the score, rounds, surface, tourney names, and countries with simple label encoding


```python
#Score
#I'll need to account for grandslams which are best of 5

#'G' = Grand Slams 
#'M' = Masters 100s
#'A' = other tour-level events
#'C' = Challengers
#'S' = Satellites/ITFs
#'F' = Tour finals and other season-ending events
#'D' = Davis Cup.

#Slams:   1 : straight sets  1.5 : 4 sets   2 : 5 sets
#other:   1 : straight sets  2 : 3 sets   

data = df.reset_index(drop=True)
data.score.apply(str)
#Score
for x in range(len(data)):
    row = data.iloc[x]    
    thing = row['score'].split(' ')
    if row['tourney_level'] == 'G': #Best of 5 e.g. Grandslams
        if len(thing) == 3:
            data.at[x, 'score'] = 1
        elif len(thing) == 4:
            data.at[x, 'score'] = 1.5
        else:
            data.at[x, 'score'] = 2
    else:   #Best of 3
        if len(thing) == 3:
            data.at[x, 'score'] = 2
        else:
            data.at[x, 'score'] = 1
            
data['score'].unique()
```

```python
#Reference table
from sklearn.preprocessing import LabelEncoder

a = data['round'].drop_duplicates()
b = data['surface'].drop_duplicates()
c = data['tourney_name'].drop_duplicates()
x = data['winner_ioc'].drop_duplicates()
#p = data['winner_name'].drop_duplicates()

categorical_cols = ['round', 'surface', 'tourney_name', 'winner_ioc']
#categorical_cols = ['round', 'surface', 'tourney_name', 'winner_ioc', 'winner_name']
le = LabelEncoder()
data[categorical_cols] = data[categorical_cols].apply(lambda col: le.fit_transform(col))

d = data['round'].drop_duplicates().rename('value')
e = data['surface'].drop_duplicates().rename('value')
f = data['tourney_name'].drop_duplicates().rename('value')
#xx = data['winner_ioc'].drop_duplicates().rename('value')


roundReference = pd.concat([a,d], axis = 1).set_index('round').reindex()
surfaceReference = pd.concat([b,e], axis = 1).set_index('surface').reindex()
tourneyReference = pd.concat([c,f], axis = 1).set_index('tourney_name').reindex()
countryReference = pd.concat([x,xx], axis = 1).set_index('winner_ioc').reindex()
#playerReference = pd.concat([p,pp], axis = 1).set_index('winner_name').reindex()
```

```python
#R/L handedness
categorical_cols = ['winner_hand', 'loser_hand']
le = LabelEncoder()
data[categorical_cols] = data[categorical_cols].apply(lambda col: le.fit_transform(col))
#1 is Right handed 0 is Left handed
```

```python
#Now I need to use the countryReference table to assign values to the loser_ioc column
for x in range(len(data)):
    country = data.iloc[x]['loser_ioc']
    try:
        value = countryReference.loc[data.iloc[x]['loser_ioc']][0]
        data.at[x, 'loser_ioc'] = value
    except:
        data.at[x, 'loser_ioc'] = 0

#Need to do the same for winner and loser columns
#for x in range(len(data)):
#    player = data.iloc[x]['loser_name']
#    try:
#        value = playerReference.loc[data.iloc[x]['loser_name']][0]
#        data.at[x, 'loser_name'] = value
#    except:
#        data.at[x, 'loser_name'] = 0
```
